```python
df = pd.read_csv(io.BytesIO(uploaded['Property  (3).csv']), encoding='latin-1')
```

```python
print(df.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
```

```python
df['Area'] = LE.fit_transform(df['Area'])
df['Dimension'] = LE.fit_transform(df['Dimension'])
df['Property Type'] = LE.fit_transform(df['Property Type'])
df['Architecture'] = LE.fit_transform(df['Architecture'])
```

```python
df
```

```python
print(df.dtypes)
```

```python
df.corr()
```

```python
df.drop(['Dimension'], axis=1, inplace=True)
df.drop(['Property Type'], axis=1, inplace=True)
df.drop(['Bedroom'], axis=1, inplace=True)
```

```python
df
```

```python
X = df[['Size','Area']]
```

```python
Y= df['Architecture']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.4,random_state=42)

X_train, X_test, y_train, y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X, Y)
```

```python
l_score = clf.score(X_test,y_test)
l_score*100
```

```python
import pickle
pickle_out = open("clf.pkl", "wb")
pickle.dump(clf, pickle_out)
pickle_out.close()
```

```python
import streamlit as st
```

```python
pickle_in = open('clf.pkl', 'rb')
clf = pickle.load(pickle_in)
```

```python
def welcome():
    return 'welcome all'

# defining the function which will make the prediction using
# the data which the user inputs
def prediction(Size,Area):

    prediction = clf.predict(
        [[Size,Area]])
    print(prediction)
    return prediction
```

```python
# this is the main function in which we define our webpage
def main():
      # giving the webpage a title
    st.title("Design Prediction")

    # here we define some of the front end elements of the web page like
    # the font and background color, the padding and the text to be displayed
    html_temp = """
    <div style ="background-color:yellow;padding:13px">
    <h1 style ="color:black;text-align:center;">Streamlit Iris Flower Classifier ML App </h1>
    </div>
    """

    # this line allows us to display the front end aspects we have
    # defined in the above code
    st.markdown(html_temp, unsafe_allow_html = True)

    # the following lines create text boxes in which the user can enter
    # the data required to make the prediction
    Size = st.text_input("Size", "Type Here")
    Area = st.text_input("Area", "Type Here")
    result =""

    # the below line ensures that when the button called 'Predict' is clicked,
    # the prediction function defined above is called to make the prediction
    # and store it in the variable result
    if st.button("Predict"):
        result = prediction(Size,Area)
    st.success('The output is {}'.format(result))

if __name__=='__main__':
    main()
```
